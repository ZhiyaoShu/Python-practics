```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bookdata=pd.read_csv('prog_book.csv')
df=pd.DataFrame(bookdata)
```

```python
sns.histplot(data=df,x='Price', kde=True)
plt.yticks([0, 20, 40, 60, 80])
```

```python
sns.countplot(data=df, x='Type')
```

```python
colors=sns.color_palette('husl',len(df['Type'].unique()))
sns.stripplot(data=df,x='Type',y='Rating',palette=colors)

```

```python
sns.boxplot(data=df,y='Rating')
```

```python
sns.displot(data=df,x='Rating',kde=True)
```

```python
sns.heatmap(df.corr())
```

```python
type_counts=df['Type'].value_counts()
plt.pie(type_counts,autopct='%1.1f%%', labels=['Paperback','Hardcover','Kindle Edition','ebook','Unknown Binding','Box Set-Hardcover'])
```
